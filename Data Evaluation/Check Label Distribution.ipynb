```python
import pandas as pd
import glob
import seaborn as sn
from tqdm import tqdm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arousal_label = []
valence_label = []
for f in tqdm(glob.glob('../local_cache/AVEC/labels/*/*.csv')):
    df = pd.read_csv(f)
    arousal_label.extend(df['Arousal'])
    valence_label.extend(df['Valence'])
```

```python
sn.distplot(arousal_label)
plt.title('Arousal Distribution')
plt.xlabel('Arousal')
plt.ylabel('Relative Distribution')
plt.show()

sn.distplot(valence_label)
plt.title('Valence Distribution')
plt.xlabel('Valence')
plt.ylabel('Relative Distribution')
plt.show()
```
